
# 선 그래프 (Line Plot)
선 그래프는 시간이나 순서에 따른 데이터의 변화 추세를 보여줄 때 유용

```python
# 데이터
months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
temperature = [0.2, 2.5, 7.8, 13.9, 19.1, 23.4, 26.8, 27.3, 22.5, 16.2, 8.8, 2.1]

# 그래프 생성
plt.figure(figsize=(12, 6))
plt.plot(months, temperature, marker='o') # 'o' 마커 추가

# 제목 및 레이블
plt.title('2024년 월별 평균 기온')
plt.xlabel('월')
plt.ylabel('기온 (°C)')

# 그리드(격자) 추가
plt.grid(True, linestyle='--')
plt.show()
```

[예제] 여러 개의 선 그래프 비교 (두 제품 판매량)
두 가지 제품의 일별 판매량을 하나의 차트에서 비교


```python
# 데이터
days = np.arange(1, 11) # 1일부터 10일
sales_A = [100, 105, 120, 115, 130, 140, 135, 150, 160, 155]
sales_B = [80, 88, 95, 100, 110, 105, 115, 120, 125, 130]

# 그래프 생성
plt.figure(figsize = (10,6))

# 제품 A: 파란색 실선, 원형 마커
plt.plot(days, sales_A, color='blue', marker='o', linestyle='-', label='제품 A')
# 제품 B: 빨간색 점선, 사각형 마커
plt.plot(days, sales_B, color='red', marker='s', linestyle='--', label='제품 B')

# 제목, 레이블, 범례
plt.title('제품 A와 B의 일별 판매량 비교')
plt.xlabel('일 (Day)')
plt.ylabel('판매량')
plt.legend(loc='upper left') # 범례 표시
plt.grid(True)
plt.xticks(days) # X축 눈금을 1부터 10까지 모두 표시
plt.show()

print(days)
```

# 산점도 (Scatter Plot)
산점도는 두 변수 간의 관계(상관관계)를 파악할 때 유용

[예제] 기본 산점도 (키와 몸무게)
학생들의 키(X축)와 몸무게(Y축) 간의 관계를 점으로 표시

```python
# 가상 데이터 생성 (학생 50명)
np.random.seed(42) # 재현 가능성을 위해 시드 고정 (42번 시나리오)
height = np.random.normal(170, 8, 50) # 평균 170cm, 표준편차 8 (노말분포 정규분포) - 3가지 인수 (평균, 표준편차, 데이터 개수)
weight = (height - 100) * 0.8 + np.random.normal(0, 5, 50) # 키와 비례 + 노이즈

print(height)
print(weight)

# 그래프 생성
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='purple', alpha=0.6) # alpha: 투명도


# 제목 및 레이블
plt.title('학생들의 키와 몸무게 분포')
plt.xlabel('키 (cm)')
plt.ylabel('몸무게 (kg)')
plt.grid(True, linestyle=':')
plt.show()
```

[예제] 그룹별 산점도 (색상 및 크기 활용)
데이터를 그룹별로 나누어 색상(c)으로 구분하고, 크기(s)를 다르게 하여 추가 정보를 표현

```python
# 가상 데이터 생성 (100개 데이터, 2개 그룹)
np.random.seed(1)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)

# 그룹 (0 또는 1)
groups = np.random.randint(0, 2, N)
# 크기 (데이터의 중요도나 다른 값)
sizes = 1000 * np.random.rand(N) # 2차원 상에서 3차원 정보를 제공 (점의 크기)

# 그래프 생성
plt.figure(figsize=(10, 7))
# c=groups : 그룹에 따라 색상 자동 지정
# s=sizes : 점마다 다른 크기 지정
# cmap='cool' : 색상 맵 지정 (0=파란색 계열, 1=붉은색 계열)
scatter = plt.scatter(x, y, c=groups, s=sizes, alpha=0.7, cmap='cool')

# 제목 및 레이블
plt.title('그룹별/크기별 데이터 분포')
plt.xlabel('X 변수')
plt.ylabel('Y 변수')

# 컬러바(Colorbar) 대신 범례(Legend) 생성
# 점의 크기가 아닌, 색상(그룹)에 대한 범례를 만듭니다.
legend1 = plt.legend(*scatter.legend_elements(), title="그룹")
plt.gca().add_artist(legend1)
plt.show()
```

# 막대 그래프 (Bar Chart)
- 용도: 범주형 데이터의 빈도나 값을 비교할 때 사용
- 주요 함수: plt.bar(x, height)
- x: 막대의 위치(보통 범주 이름)
- height: 막대의 높이(보통 해당 범주의 값)

[예제] 5개의 범주를 막대그래프로 표현하여 크기를 비교

```python
# 1. 데이터 준비
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 10]

# 2. 막대 그래프 생성
plt.figure(figsize=(8, 5)) # 그래프 크기 설정
plt.bar(categories, values, color='skyblue')

# 3. 그래프 제목 및 축 레이블 설정
plt.title('카테고리별 값 비교')
plt.xlabel('카테고리')
plt.ylabel('값')

# 4. 그래프 출력
plt.grid(axis='y', linestyle='--') # Y축에 가이드 라인 추가
plt.show()
```

# 히스토그램 (Histogram)
용도: 연속적인 데이터의 분포를 파악할 때 사용. 데이터를 여러 개의 구간(bin)으로 나누어 각 구간에 속하는 데이터의 빈도를 막대로 나타냄.
주요 함수: plt.hist(x, bins)
x: 분포를 보고 싶은 데이터 배열
bins: 구간(bin)의 개수 또는 경계 (생략 시 기본값 사용)

막대기가 붙어있고 x축이 범주가 아닌 숫자면 히스토그램.

[예제] 무작위생성 데이터 1000개의 히스토그램 작성

```python
# 1. 데이터 준비 (랜덤하게 정규 분포 데이터 1000개 생성)
data = np.random.randn(1000)

# 2. 히스토그램 생성
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='lightcoral', edgecolor='black') # 30개의 bin 설정

# 3. 그래프 제목 및 축 레이블 설정
plt.title('랜덤 데이터의 분포 (히스토그램)')
plt.xlabel('값')
plt.ylabel('빈도수')

# 4. 그래프 출력
plt.show()
```

# 파이 차트 (Pie Chart)
용도: 전체에 대한 각 부분의 비율을 나타낼 때 사용
주요 함수: plt.pie(x, labels, autopct)
x: 각 부분의 크기(비율)를 나타내는 값 배열
labels: 각 부분의 레이블(이름)
autopct: 비율을 표시하는 형식 (예: '%.1f%%'는 소수점 첫째 자리까지 표시)

[예제] 과일 선호도에 대하여 100명에게 물어본 설문조사의 결과를 파이차트로 작성

```python
# 1. 데이터 준비
sizes = [15, 30, 45, 10]
labels = ['사과', '바나나', '오렌지', '포도']
colors = ['yellowgreen', 'gold', 'lightcoral', 'lightskyblue']

# 2. 파이 차트 생성
plt.figure(figsize=(7, 7))
plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%', # 소수점 첫째 자리까지 비율 표시
        startangle=90, # 90도에서 시작 (위에서부터)
        shadow=True)

# 3. 그래프 제목 설정
plt.title('과일 선호도 비율')

# 4. 파이 차트가 원형으로 보이도록 설정
plt.axis('equal')

# 5. 그래프 출력
plt.show()
```
