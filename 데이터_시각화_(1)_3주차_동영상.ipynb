꺾은선 그래프 (Line Plot) 3개 예제

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비 (웹사이트 트래픽)
days = np.arange(1,11)
traffic = [1200, 1350, 1300, 1420, 1550, 1500, 1600, 1750, 1700, 1850]

# 그래프 생성
plt.figure(figsize = (10, 5)) #가로 10, 세로 5

# 스타일을 지정하여 라인 플롯 그리기
plt.plot(days,traffic,      # x, y 데이터
         color = 'blue',    # 색깔
         marker = 'o',      # 마커 모양: 원형
         markersize = 8,     # 마커 크기
         linestyle = '--',  # 선 스타일: 점선 (dashed)
         linewidth = 2)     # 선 두께

# 그래프 꾸미기
plt.title("일별 웹사이트 트래픽")
plt.xlabel("날짜")
plt.ylabel("방문자수(명)")
plt.xticks(days)
plt.grid(True)

# 그래프 표시
plt.show()
```

- (행, 열)의 격자 형태로 그래프를 배치할 수 있음.
