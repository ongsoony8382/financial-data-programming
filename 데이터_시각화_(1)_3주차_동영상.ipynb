꺾은선 그래프 (Line Plot) 3개 예제

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비 (웹사이트 트래픽)
days = np.arange(1,11)
traffic = [1200, 1350, 1300, 1420, 1550, 1500, 1600, 1750, 1700, 1850]

# 그래프 생성
plt.figure(figsize = (10, 5)) #가로 10, 세로 5

# 스타일을 지정하여 라인 플롯 그리기
plt.plot(days,traffic,      # x, y 데이터
         color = 'blue',    # 색깔
         marker = 'o',      # 마커 모양: 원형
         markersize = 8,     # 마커 크기
         linestyle = '--',  # 선 스타일: 점선 (dashed)
         linewidth = 2)     # 선 두께

# 그래프 꾸미기
plt.title("일별 웹사이트 트래픽")
plt.xlabel("날짜")
plt.ylabel("방문자수(명)")
plt.xticks(days)
plt.grid(True)

# 그래프 표시
plt.show()
```

[예제] 서브플롯 (Subplots) 사용하기

- 여러 개의 그래프를 독립된 영역에 나누어 그리고 싶을 때 plt.subplots()를 사용함.
- (행, 열)의 격자 형태로 그래프를 배치할 수 있음.
- 여기서는 2행 1열 구조로 두 개의 그래프를 그림.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100) # 0부터 10까지 100개의 점
y1 = np.sin(x) # sin 곡선
y2 = np.cos(x) # cos 곡선

# 2행 1열의 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (8,8)) # (행, 열, 크기)

# 첫 번째 그래프 (ax1)
ax1.plot(x, y1, color = 'blue', label = 'sin(x)')
ax1.set_title('Sine Wave')
ax1.set_ylabel('sin(x)')
ax1.grid(True)
ax1.legend()

# 두 번째 그래프 (ax2)
ax2.plot(x, y2, color = 'red', label = 'cos(x)')
ax2.set_title('Cosine Wave')
ax2.set_ylabel('cos(x)')
ax2.grid(True)
ax2.legend()

# 레이아웃 자동 조절
plt.tight_layout()

# 그래프 표시
plt.show()
```


[예제] Y축 범위 및 로그 스케일 지정
데이터의 편차가 매우 크거나 지수적으로 증가하는 경우, Y축의 범위를 강제로 지정하거나(plt.ylim()) 로그 스케일(plt.yscale('log'))을 사용하면 데이터의 패턴을 더 명확하게 볼 수 있음.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비 (지수적 증가)
x = np.arange(1,10)
y = np.exp(x) # 지수 함수 (값이 매우 빠르게 커짐)

# 1행 2열 서브플롯 (일반 스케일 vs 로그 스케일)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5)) #1행 2열, 가로 12 세로 5

# 1. 일반 스케일 (Linear Scale)
ax1.plot(x, y, marker = 'o')
ax1.set_title('Linear Scale')
ax1.set_xlabel('x')
ax1.set_ylabel('exp(x)')
ax1.grid(True)

# 2. 로그 스케일 (Log Scale)
ax2.plot(x, y, marker = 's', color = 'orange')
ax2.set_yscale('log') # y축을 로그 스케일로 변경
ax2.set_title('Log Scale')
ax2.set_xlabel('x')
ax2.set_ylabel('exp(x) (log scale)')
ax2.grid(True, which = 'both') # 주/보조 그리드 모두 표시

# 그래프 표시
plt.show()
```

산점도 (Scatter Plot) 3개의 예제
산점도는 두 개의 서로 다른 숫자 변수 간의 관계(상관관계)를 파악하는 데 사용.


[예제] 세 번째 변수를 색상으로 표현하기 (Color Mapping)
- 산점도는 기본적으로 2차원(X, Y)이지만, 점의 색상을 이용해 3차원 변수를 표현할 수 있음.
- c 인자에 Y값이 아닌 다른 배열을 전달하고 cmap (컬러맵)을 지정.
- plt.colorbar()로 색상 막대를 추가


```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
np.random.seed(42)
N = 100
x = np.random.randn(N) # X (정규분포) normal distribution
y = np.random.randn(N) # Y (정규분포) normal distribution
colors = np.random.rand(N) # 0~1 사이의 랜덤값

# 그래프 생성
plt.figure(figsize = (9,6))

# colors 변수를 c 인자에 전달. viridis 컬러맵 사용
scatter = plt.scatter(x, y, c = colors, cmap = 'viridis', alpha = 0.8)

# 컬러바 (색상 막대) 추가
plt.colorbar(scatter, label = '세 번째 변수 (Color Value)')

# 그래프 꾸미기
plt.title("세 번째 변수를 색상으로 표현한 산점도")
plt.xlabel("x 변수")
plt.ylabel("y 변수")
plt.grid(True, linestyle =':', alpha = 0.5)

# 그래프 표시
plt.show()
```

# [예제] 네 번째 변수를 크기로 표현하기 (Size Mapping)
색상뿐만 아니라 점의 크기(s)를 조절하여 네 번째 변수를 표현할 수도 있음.
이를 버블 차트(Bubble Chart)라고도 부름.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
np.random.seed(42)
N = 50
x = np.random.rand(N) # X (제품 만족도)
y = np.random.rand(N) # Y (제품 가격)
colors = np.random.rand(N) # 세 번째 변수 (색상: 재구매 의사)
sizes = 1000 * np.random.rand(N) # 네 번째 변수 (크기: 판매량)

# 그래프 생성
plt.figure(figsize=(10, 7))

# 크기(s)와 색상(c)을 모두 다르게 지정
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='plasma')

# 그래프 꾸미기
plt.title('버블 차트 (크기와 색상으로 4개 변수 표현)')
plt.xlabel('제품 만족도 (0-1)')
plt.ylabel('제품 가격 (0-1)')
plt.grid(True, linestyle=':', alpha=0.5)

# (참고: 크기와 색상에 대한 범례는 별도 처리가 필요할 수 있음)

# 그래프 표시
plt.show()
```

# [예제] 그룹별로 다른 마커/색상 사용하기
데이터가 여러 그룹(범주)으로 나뉘어 있을 때, 각 그룹별로 다른 색상이나 마커를 사용하여 시각적으로 구분할 수 있음.
예를 들어, 붓꽃(iris) 데이터의 세 가지 품종을 구분하는 경우.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비 (가상의 3개 그룹 데이터)
np.random.seed(10)
# 그룹 A
group_a_x = np.random.normal(2, 1, 30)
group_a_y = np.random.normal(2, 1, 30)
# 그룹 B
group_b_x = np.random.normal(5, 1, 30)
group_b_y = np.random.normal(5, 1, 30)
# 그룹 C
group_c_x = np.random.normal(3, 1, 30)
group_c_y = np.random.normal(8, 1, 30)

# 그래프 생성
plt.figure(figsize=(9, 7))

# 각 그룹별로 산점도 그리기
plt.scatter(group_a_x, group_a_y, c='red', marker='o', label='그룹 A', alpha=0.7)
plt.scatter(group_b_x, group_b_y, c='blue', marker='^', label='그룹 B', alpha=0.7)
plt.scatter(group_c_x, group_c_y, c='green', marker='s', label='그룹 C', alpha=0.7)

# 그래프 꾸미기
plt.title('그룹별로 구분된 산점도')
plt.xlabel('X 특성')
plt.ylabel('Y 특성')
plt.legend() # 범례 표시
plt.grid(True, linestyle=':')

# 그래프 표시
plt.show()
```
