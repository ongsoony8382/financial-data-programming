# [예제] Plotly 인터랙티브 시각화: 레스토랑 팁(Tips)
**tips 데이터셋**
- 그룹별(categorical) 비교에 아주 좋은 예제

- 한 레스토랑에서 손님들이 지불한 팁에 대한 정보를 담고 있으며, 다음과 같은 컬럼을 포함

- total_bill: 총 식사 금액
- tip: 팁 금액
- sex: 성별 (Male / Female)
- smoker: 흡연자 여부 (Yes / No)
- day: 요일 (Sun, Sat, Thur, Fri)
- time: 시간대 (Dinner / Lunch)
- size: 일행 수

이 데이터를 사용해 다양한 그룹 간의 차이를 인터랙티브하게 탐색


# 라이브러리와 데이터 불러오기
- 먼저 Colab에서 Plotly 라이브러리와 tips 데이터를 불러옴.

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# import plotly
# plotly.offline.init_notebook_mode(connected=True)

# plotly에 내장된 'tips' 데이터셋 로드
df = px.data.tips()

# 데이터가 어떻게 생겼는지 확인 (상위 5개행)
print(df.head())
```

# 산점도 (Scatter Plot): 식사 금액과 팁의 관계
gapminder에서 'GDP'와 '기대 수명'의 관계를 보았듯이, tips에서는 '총 식사 금액'과 '팁'의 관계를 볼 수 있음.

- x축: total_bill (총 식사 금액)
- y축: tip (팁 금액)
- 색상: sex (성별)
- 마커 크기: size (일행 수)
- 패널 분리: smoker (흡연 여부)

```python
fig = px.scatter(df,
                 x='total_bill',
                 y='tip',
                 color='sex', # 성별에 따라 색상 구분
                 size='size', # 일행 수에 따른 원 크기
                 facet_col='smoker', # 흡연 여부에 따라 그래프 패널 분리
                 hover_name='day', # 마우스 올리면 요일 표시
                 title='총 식사 금액과 팁의 관계 (성별/일행 수/흡연 여부별)')
fig.show()
```

# 인터랙티브 탐색:
- 오른쪽 범례에서 'Male' 또는 'Female'을 클릭해 보자.
- 흡연자(Yes) 그룹과 비흡연자(No) 그룹의 total_bill과 tip 분포가 시각적으로 다른가?

# 2) 막대 그래프 (Bar Chart): 요일별 팁 합계 (성별 비교)
- gapminder에서 '대륙별 인구'를 비교했다면, 여기서는 '요일별 팁'을 비교할 수 있음.
- barmode='group'을 사용하면 그룹별 비교가 더 쉬워짐.

```python
fig = px.bar(df,
             x='day',
             y='tip',
             color='sex',
             barmode='group', # group 모드로 막대를 옆으로 나란히  표시
             title="요일별/성별 팁 합계 비교")

fig.show()
```

# 인터랙티브 탐색:
- 'Sat'(토요일)와 'Sun'(일요일)이 다른 요일보다 팁 합계가 높은 것을 볼 수 있다.
- 그래프 위에 마우스를 올려 특정 막대의 정확한 값을 확인해 보자.

# 박스 플롯 (Box Plot): 시간대별 팁 분포
- 박스 플롯은 데이터의 중앙값, 사분위수, 이상치(outlier)를 한눈에 보여줘 그룹 간 분포를 비교할 때 매우 강력함.

- x축: time (시간대: Lunch / Dinner)
- y축: tip (팁 금액)
- 색상: smoker (흡연 여부)

```python
fig = px.box(df,
             x='time',
             y='tip',
             color='smoker',
             title="시간대별/흡연여부별 팁 분포 비교",
             points='all') # all로 설정하면 모든 데이터 포인트를 같이 표시
fig.show()
```

# 인터랙티브 탐색:
'Dinner' 시간대의 팁 중앙값(상자 안의 선)이 'Lunch'보다 높은 것을 알 수 있음.
각 그룹에서 유난히 팁을 많이 낸 이상치(다른 점들과 멀리 떨어진 점)를 확인해 보자.

# 히스토그램 (Histogram): 총 식사 금액의 분포 (성별 비교)

- 데이터가 어떻게 분포되어 있는지 확인할 때 히스토그램을 사용
- marginal='rug'를 추가하면 데이터의 실제 위치를 함께 볼 수 있음.

```python
fig = px.histogram(df,
                   x='total_bill',
                   color='sex',
                   nbins=30, # 막대 개수 설정
                   marginal='rug', #'rug' plot을 상단에 추가해 실제 데이터 위치 표시)
                   title='총 식사 금액 분포 (성별비교)')

fig.show()
```

# [예제] Plotly 인터랙티브 시각화: 몬트리올 카셰어링 핫스팟 찾기
이 예제는 지도 위에 점(scatter)을 찍어, 어느 위치에서, 어느 시간대에, 얼마나 오래 카셰어링이 이용되었는지 한눈에 보여줌.

# 데이터 준비
먼저 carshare 데이터를 불러옴.

```python
import plotly.express as px
import pandas as pd

# 'carshare' 데이터셋 로드
df = px.data.carshare()

# 데이터 확인 (위도,, 경도, 이용 시간, 피크 시간)
print(df.head())

# latitude, longitude
```

# 지도에 산점도 그리기 (Scatter Mapbox)
px.scatter_mapbox() 함수를 사용.

이 함수는 Plotly의 맵박스(Mapbox) 서비스와 연동되어 매우 부드럽고 빠른 인터랙티브 지도를 제공함.

lat, lon: 위도, 경도 컬럼을 지정
color: 'peak_hour' (피크 시간, 0~23시)에 따라 색상을 다르게 표시
size: 'car_hours' (이용 시간)에 따라 원의 크기를 다르게 표시.
map_style: 지도의 배경 스타일을 지정 ("open-street-map", "carto-positron" 등)
zoom: 지도의 초기 확대/축소 레벨을 설정

```python
fig = px.scatter_mapbox(df,
                       lat="centroid_lat",      # 위도
                       lon="centroid_lon",      # 경도
                       color="peak_hour",       # 색상: 피크 시간 (0~23시)
                       size="car_hours",        # 크기: 이용 시간
                       hover_name="car_hours",  # 마우스 올리면 이용 시간 표시

                       color_continuous_scale=px.colors.cyclical.IceFire, # 색상 테마
                       size_max=15,             # 원 최대 크기
                       zoom=10,                 # 초기 줌 레벨 (몬트리올 시내가 보이도록)
                       mapbox_style="carto-positron", # 맵 스타일 (밝은 배경)
                       title="몬트리올 카셰어링 이용 현황 (시간대별/이용 시간별)")

fig.show()
```

# 실제 지도 탐색
# **줌 & 패닝 (Zoom & Panning):**
- 차트가 로드되면 마우스 휠을 스크롤하여 지도를 확대/축소해보자.
지도를 클릭하고 드래그하여 몬트리올 시내 곳곳을 탐색할 수 있다.
- scatter_mapbox는 줌 레벨에 따라 지도의 상세 정보(도로명 등)가 자동으로 업데이트됨.
# **색상으로 시간대 파악:**
- 오른쪽의 색상 막대(Color Bar)는 0시(자정)부터 23시(밤 11시)까지의 peak_hour를 나타냄.
- 특정 지역에 어떤 색상의 점들이 몰려있는지 확인해보자. (예: 주거 지역은 밤 시간대, 오피스 지역은 낮 시간대)
# **크기로 이용 시간 파악:**
- 원이 클수록(size='car_hours') 더 장시간 이용되었다는 뜻.
어느 지역에서 차량이 더 오래 사용되는지 한눈에 파악할 수 있음.
